```python
# Shorthand for printing both name and value of an expression

long_var_name = 1.2
a = 3.4 
b = 5.6

print("<< long_var_name =", long_var_name, ">>")    # without using f-string
print(f"<< long_var_name = {long_var_name} >>")     # using f-string
print(f"<< {long_var_name = } >>")                  # using f-string shorthand

print()
print(f"a + b + 1 =", a + b + 1)
print(f"a + b + 1 = {a + b + 1}")
print(f"{a + b + 1 = }")

print()
print(f"{max(a, b) = }")
print(f"{max(5, 6) = }")
```

```python
# Text alignment and padding

txt = "hello"
print(f"|{txt}|")       # no padding
print(f"|{txt:10}|")    # def alignment = left, def padding = space
print(f"|{txt:>10}|")   # right align
print(f"|{txt:^10}|")   # center align
print(f"|{txt:_<10}|")  # left align with custom padding
print(f"|{txt:=^10}|")  # center align with custom padding
```

```python
# Number separator, precision, alignment and padding

n = 1_000_000_000 # use underscore for better readability
print(f"{n}")
print()

n = 1000000000
print(f"{n:_}") # underscore can be used as separator
print(f"{n:,}") # comma can be used as separator
print()

n = 123.456
print(f"{n:.2f}") # for under precision round offs to 2 decimal places
print(f"{n:.4f}") # for over precision adds trailing zeroes to 4 decimal places
print()

n = 1234.5678
print(f"|{n}|")         # no padding
print(f"|{n:20.2f}|")   # def alignment = left, def padding = space
print(f"|{n:<20.2f}|")  # left align
print(f"|{n:^20.2f}|")  # center align
print(f"|{n:0<20.2f}|") # left align with 0 padding
print(f"|{n:0>20.2f}|") # right align with 0 padding
print(f"|{n:0^20.2f}|") # center align with 0 padding
print()

print(f"{n:,f}")
print(f"{n:,.2f}")
```

```python
# Date and time format specifiers
from datetime import datetime

now = datetime.now()
print(f"{now:%x}")      # %x is the default date format
print(f"{now:%X}")      # %X is the default time format
print(f"{now:%c}\n")    # %c is the default date and time format

print(f"Hour (24-hour format) : {now:%H}")   # %H is the hour (24-hour format)
print(f"Hour (12-hour format) : {now:%I}")   # %I is the hour (12-hour format)
print(f"AM/PM indicator       : {now:%p}\n") # %p is the AM/PM indicator

print(f"Minute                : {now:%M}")   # %M is the minute
print(f"Second                : {now:%S}")   # %S is the second
print(f"Microsecond           : {now:%f}")   # %f is the microsecond
```

```python
# Date and time format specifiers
from datetime import datetime

now = datetime.now()
print(f"Day of the month      : {now:%d}")   # %d is the day of the month
print(f"Day of the year       : {now:%j}\n") # %j is the day of the year

print(f"Week of the year (Sn) : {now:%U}")    # %U is the week number with Sunday as the first day of week
print(f"Week of the year (Mn) : {now:%W}")   # %W is the week number with Monday as the first day of week
print(f"Weekday number        : {now:%w}")   # %w is the weekday number
print(f"Weekday name abbr     : {now:%a}")   # %A is the weekday name
print(f"Weekday name full     : {now:%A}\n") # %A is the weekday name

print(f"Month number          : {now:%m}")   # %m is the month number
print(f"Month name abbr       : {now:%b}")   # %b is the month abbreviation
print(f"Month name full       : {now:%B}\n") # %B is the month name

print(f"Year without century  : {now:%y}")   # %y is the year without century
print(f"Year with century     : {now:%Y}\n") # %Y is the year with century
```
