## Scatter plot

- `scatter()` is a method in `matplotlib.pyplot` used to create scatter plots.
- this shows individual data points (dots) in a 2D space.
- Unlike `plot()`, it does not connect the dots with lines.

use cases:
- Data exploration: Scatter plots help find patterns, clusters, outliers, or relationships between two variables.
- Comparing variables: E.g., height vs weight, age vs income, etc.
- Bubble charts (using different sized dots).

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.show()
```

```python
x = [1, 2, 3, 5]
y1 = [10, 20, 25, 30]
y2 = [15, 20, 15, 10]

plt.scatter(x, y1, label='First Set', color='green', marker='o')
plt.scatter(x, y2, label='Second Set', color='red', marker='x')

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.xlim(0, 10)
plt.xticks([1, 2, 3, 4, 5, 6])
plt.ylim(0, 40)
plt.yticks([0, 10, 20, 30, 40, 50])
plt.title('My Title')
plt.grid(visible=True, axis='both', linestyle='--')
plt.legend()
plt.show()

# other valid values
# markers: 'o', 'v', '^', '<', '>', '*', '+', 'x', 'X', 'D', 'd', 's', 'p', 'h', 'H'
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 4, 5]
colors = [20, 10, 30, 45, 50]
sizes = [100, 200, 300, 400, 500]

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, edgecolors='black')
plt.colorbar()  # Adds a color scale bar
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

## Line plot

## Error Bars

- An `error bar` is a graphical representation of the variability or uncertainty of data.
- It typically shows as lines above and below (or left and right of) a data point.

use cases:
- Experimental science: Representing measurement errors
- Machine learning: Visualizing confidence intervals
- Business forecasting: Showing prediction ranges

```python
x = [1, 2, 3, 4, 5]
y = [2.5, 3.0, 2.0, 4.0, 3.5]
xerr = [[0.1, 0.2, 0.15, 0.2, 0.1],   # lower errors
        [0.2, 0.1, 0.2, 0.3, 0.2]]   # upper errors
yerr = [[0.1, 0.2, 0.15, 0.2, 0.1],   # lower errors
        [0.2, 0.1, 0.2, 0.3, 0.2]]   # upper errors

plt.errorbar(x, y, xerr=xerr, yerr=yerr, color='green', ecolor='red', elinewidth=1, capsize=5)
plt.title("Error Bar Example")
plt.xlabel("X Value")
plt.ylabel("Measured Value")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 100)
y =np.sin(x)

plt.errorbar(x, y, xerr= 1, yerr=2)
plt.grid(True)
plt.show()
```

## Pie chart

use cases:

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 2, 1, 4]

plt.pie(values, labels=categories)
plt.show()
```

## Bar graph

- `bar()` function in `matplotlib.pyplot` is used to create vertical bar charts.
- The height of each bar in a `bar graph` represents a value or count corresponding to a category.

use cases:
- Category comparison: Ideal for comparing discrete categories (e.g., sales by product, frequency of responses).
- Counts/frequencies: Show how often something occurs.

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 2, 1, 4]

# for vertical bars
plt.bar(categories, values)
plt.show()

# for horizontal bars
plt.barh(categories, values)
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [100, 150, 90, 120]
colors = ['red', 'yellow', 'orange', 'brown']
labels = ['A', 'B', 'C', 'D']

plt.bar(fruits, sales, width=0.6, color=colors, edgecolor='black', linewidth=1.5, label=labels)

for i, v in enumerate(sales):
    plt.text(i, v+2, str(v), ha='center')

plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.yticks(np.arange(max(sales) + 20, step=20))
plt.legend()
plt.show()
```

## Double bar graph

- A `double bar graph` displays two bars per category
- It allows you to compare two metrics across categories (e.g., budget vs actual, this year vs last year)

use cases:
- Compare two metrics across categories (e.g., budget vs actual, this year vs last year)
- Analyze male vs female, before vs after, A vs B in controlled experiments
- Any case where you want to show grouped categorical comparison

```python
labels = ['Jan', 'Feb', 'Mar', 'Apr']
product_A = [10, 15, 7, 12]
product_B = [12, 17, 6, 10]

x = np.arange(len(labels))  # [0, 1, 2, 3]
width = 0.3  # Width of each bar

plt.bar(x - width/2, product_A, width, label='Product A', color='orange')
plt.bar(x + width/2, product_B, width, label='Product B', color='brown')

for i in range(len(x)):
    plt.text(x[i] - width/2, product_A[i] + 0.2, str(product_A[i]), ha='center')
    plt.text(x[i] + width/2, product_B[i] + 0.2, str(product_B[i]), ha='center')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.xticks(x, labels)
plt.title('Monthly Sales Comparison')
plt.yticks(np.arange(max(product_A + product_B) + 3, step=3))
plt.legend()

plt.show()
```

## Tornado graph

- A `tornado graph` is a type of horizontal bar chart used to compare two opposing sets of data.
- In this, the bars extend in opposite directions from a central axis

use cases:
- Sensitivity Analysis (commonly in risk management or project evaluation)
- Population pyramids (e.g., age groups of males vs females)
- Side-by-side comparison of categories (e.g., current vs previous values)

```python
age_groups = ['0-10', '11-20', '21-30', '31-40', '41-50']
male_pop = np.array([5, 7, 6, 4, 3])    
female_pop = np.array([4, 6, 5, 3, 2])  

y = np.arange(len(age_groups))  # [0, 1, 2, 3, 4]
plt.barh(y, male_pop * (-1), color='skyblue', label='Male') # negative for left bars
plt.barh(y, female_pop * (+1), color='pink', label='Female') # positive for right bars

plt.yticks(y, age_groups)
plt.xlabel('Population (in thousands)')
plt.ylabel('Age groups')
plt.title('Population Pyramid')
plt.legend()
plt.axvline(0, color='black')
plt.show()
```

## Stacked bar graph

- A `stacked bar graph` is a type of bar chart where multiple data series are stacked on top of each other.

use cases:
- Breakdown of expenses (rent, food, travel) by month.
- Visualize component breakdowns (using percent-stacked bars).
- Population by age group and gender.

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [10, 15, 20, 25]
product_B = [5, 10, 15, 10]
product_C = [2, 3, 4, 8]
x = np.arange(len(labels)) # [0, 1, 2, 3]

plt.bar(x, product_A, label='Product A', color='skyblue')
# Adding B on top of A
plt.bar(x, product_B, bottom=product_A, label='Product B', color='salmon')
# Adding C on top of A+B
product_AB = [a + b for a, b in zip(product_A, product_B)]
plt.bar(x, product_C, bottom=product_AB, label='Product C', color='lightgreen')

plt.xticks(x, labels)
plt.ylabel('Sales')
plt.title('Quarterly Sales by Product')
plt.legend()
plt.show()
```

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [10, 15, 20, 25]
product_B = [5, 10, 15, 10]
product_C = [2, 3, 4, 8]

# normalized values
for i in range(len(product_A)):
    total = product_A[i] + product_B[i] + product_C[i]
    product_A[i] = 100 * product_A[i] / total
    product_B[i] = 100 * product_B[i] / total
    product_C[i] = 100 * product_C[i] / total

x = np.arange(len(labels))

plt.bar(x, product_A, label='Product A', color='skyblue')
# Adding B on top of A
plt.bar(x, product_B, bottom=product_A, label='Product B', color='salmon')
# Adding C on top of A+B
product_AB = [a + b for a, b in zip(product_A, product_B)]
plt.bar(x, product_C, bottom=product_AB, label='Product C', color='lightgreen')

plt.xticks(x, labels)
plt.ylabel('Percentage Sales')
plt.title('Quarterly Sales by Product')
plt.show()
```

## Stack plot

- `stackplot()` creates a stacked area plot.
- It is similar to a stacked bar chart, but with continuous data plotted as filled areas under curves, one on top of the other.

use cases:
- Cumulative data over time
- Contribution of each category to a whole across a continuous variable (like time)
- Resource usage (CPU, memory, bandwidth by process)
- Cumulative financial or population data
- Stacked area visualization in reports or dashboards

```python
days = [1, 2, 3, 4, 5]
apples = np.array([2, 3, 4, 4, 3])
bananas = np.array([5, 6, 4, 3, 4])
cherries = np.array([1, 1, 2, 3, 4])

plt.stackplot(days, apples, bananas, cherries, labels=['Apples', 'Bananas', 'Cherries'], colors=['#ff9999', '#ffe066', '#a1cfff'])
plt.legend(loc='upper left')
plt.title('Fruit Consumption Over Days')
plt.xlabel('Day')
plt.ylabel('Quantity')
plt.xticks(days)
plt.yticks([i for i in range(max(apples + bananas + cherries) + 1)])
plt.grid()
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 150)
y1 = np.cos(x)**2
y2 = np.sin(x)**2

plt.stackplot(x, y1, y2, colors=['#123456', '#abcdef'])
plt.xticks(days)
plt.grid()
plt.tight_layout()
plt.show()
```

## Fill between

- `fill_between()` fills the area between two y-values across a range of x-values.

use cases:
- Show the range of values (e.g. min to max)
- Visualize confidence intervals
- Highlight differences between two curves
- Fill area under a curve for emphasis

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, color='skyblue')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
mean = np.sin(x)
std = 0.2
plt.plot(x, mean, label='Mean')
plt.fill_between(x, mean - std, mean, alpha=0.3, label='-std')
plt.fill_between(x, mean, mean + std, alpha=0.3, label='+std')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.fill_between(x, y1, y2, where=(y1 > y2), color='green', alpha=0.3, label='sin > cos')
plt.fill_between(x, y1, y2, where=(y1 <= y2), color='red', alpha=0.3, label='cos > sin')

plt.title("Difference between sin(x) and cos(x)")
plt.legend()
plt.show()
```

## ECDF plot

An ECDF (Empirical Cumulative Distribution Function) graph plots the count (or proportion) of observations less than or equal to a particular value.

Benefits of histograms:
- optimal for grouped data
- optimal for finding mode and skewness
- Discrete graphs

Benefits of ECDF plot:
- No binning required
- Shows every individual data point
- Continuous graphs

use cases:
- Percentile checking
- Probability distribution
- Cumulative Frequency Graphs
- Distribution comparison: Easily compare multiple datasets.
- Detecting outliers: Identify jumps or flat regions.

```python
# Generate sample data
data = np.random.normal(loc=50, scale=20, size=100)
# Sort data
x = np.sort(data)
# Compute ECDF values (proportion)
y = np.arange(1, len(x)+1) / len(x)

# Plot ECDF
plt.plot(x, y, marker='None', linestyle='-')
plt.xlabel("Value")
plt.ylabel("ECDF")
plt.xticks(range(0, 101, 10))
plt.title("Empirical CDF")
plt.grid(True)
plt.show()
```

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(55, 20, 100)

x1 = np.sort(data1)
y1 = np.arange(1, len(x1)+1) / len(x1)

x2 = np.sort(data2)
y2 = np.arange(1, len(x2)+1) / len(x2)

plt.plot(x1, y1, label="Dataset 1")
plt.plot(x2, y2, label="Dataset 2")
plt.xlabel("Value")
plt.ylabel("ECDF")
plt.title("Comparing Two ECDFs")
plt.legend()
plt.grid(True)
plt.show()
```

## Box plot

A `box plot` summarizes a dataset using five key statistics:
- Minimum (0th percentile)
- First Quartile (25th percentile)
- Median (50th percentile)
- Third Quartile (75th percentile)
- Maximum (100th percentile)

In the a typical box plot:
- The `box` expands from Q1 to Q3 (interquartile range, IQR).
- The `median` is represented by a line inside the box.
- The `whiskers` extend to the min and max values within 1.5×IQR.
- `Outliers` (beyond 1.5×IQR) are shown as individual points.

use cases:
- Detect outliers in data
- Identify skewness
- Summarize large datasets without plotting all data points

```python
data = np.random.normal(loc=50, scale=10, size=1000)

plt.boxplot(data, vert=True)
plt.yticks(range(0, 101, 10))
plt.show()
```

## Violin plot

- A `violin plot` shows a kernel density estimate (KDE, sooth curve) of the data on both sides.
- A box plot in the center: showing median, quartiles, and optionally outliers.
- This lets you see both summary statistics and distribution shape in one chart.

use cases:
- Better visualizing distribution shape than a box plot
- Comparing multiple group distributions
- Showing multimodal (having multiple dense points) or skewed data clearly

```python
data = np.random.normal(50, 10, 1000)
plt.violinplot(data)
plt.show()
```

```python
# violin plot vs box plot
d1 = list(np.random.normal(75, 15, 500))
d2 = list(np.random.normal(20, 10, 500))

data = np.array(d1 + d2)

plt.violinplot(data)
plt.boxplot(data)
plt.show()
```

## Contour Graph

- A contour plot draws lines (contours) that connect points of equal value.
- A filled contour plot (contourf) shades the regions between these lines.

use cases:
- Topographic maps (elevation contours)
- Heat/distribution maps (temperature, pressure, concentration)
- Visualizing solutions of 2D functions *z = f ( x , y )*
- Decision boundaries in machine-learning feature spaces

```python
x = np.linspace(0, 10, 200)
y = np.linspace(0, 10, 200)
X, Y = np.meshgrid(x, y)

Z = ((X**2) + 2*(Y**2))

cs = plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.clabel(cs, inline=True, fontsize=8)  # annotate contour lines
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.colorbar(cs, label='Value')
plt.show()
```

```python
x = np.linspace(0, 10, 200)
y = np.linspace(0, 10, 200)
X, Y = np.meshgrid(x, y)

Z = ((X**2) + 2*(Y**2))

cs = plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.clabel(cs, inline=True, fontsize=8)  # annotate contour lines
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.colorbar(cs, label='Value')
plt.show()
```

## Quiver Graph

- A quiver plot shows arrows at points on a grid.
- The direction of each arrow represents the vector's direction.
- The length (and optionally color) represents its magnitude.

use cases:
- Physics: Visualize electric or magnetic fields
- Engineering: Show stress/strain, airflow patterns
- Meteorology: Wind speed and direction on a map
- Math/CS: Gradient fields, vector fields
- Robotics: Motion planning, vector navigation

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

U = X + Y
V = X - Y

# Plot quiver
plt.quiver(X, Y, U, V)
plt.axis('equal')  # Keep aspect ratio square
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Define vector components (e.g., rotation field)
U = -Y
V = X
magnitude = np.sqrt(U**2 + V**2)

# Plot quiver
plt.quiver(X, Y, U, V, magnitude, cmap='inferno')
plt.colorbar(label='Vector Magnitude')
plt.title("Vector Field: Rotation")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')  # Keep aspect ratio square
plt.grid(True)
plt.show()
```
